```python
import pandas as pd
```

# Importando os datasets

```python
df1 = pd.read_csv('AMEACA.csv', sep=';', encoding = 'utf8')
df2 = pd.read_csv('CALUNIA__INJURIA_E_DIFAMACAO.csv', sep=';', encoding = 'utf8')
df3 = pd.read_csv('ESTUPRO.csv', sep=';', encoding = 'utf8')
df4 = pd.read_csv('FEMINICIDIO.csv', sep=';', encoding = 'utf8')
df5 = pd.read_csv('LESAO_CORPORAL.csv', sep=';', encoding = 'utf8')
```

# Excluindo colunas diferentes

```python
df1=df1.drop(columns=['UF'])
df3=df3.drop(columns=['UF'])
df4=df4.drop(columns=['UF'])
df5=df5.drop(columns=['UF'])
df5=df5.drop(columns=['NATUREZA'])
```

# Juntando os arquivos

```python
frames = [df1, df2, df3, df4, df5]
result = pd.concat(frames)
```

```python
result
```

# Salvando dataset único

```python
result.to_csv('completo.csv', index=False, encoding='utf8')
```
